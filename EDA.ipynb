Exploratory Data Analysis (EDA)

```python
import pandas as pd
dataset = pd.read_csv("cleaned_dataset.csv")
dataset.head()
```

Checking trends over time

```python
import matplotlib.pyplot as plt
selectedCols = ['datetime', 'temperature_2m_mean', 'apparent_temperature_mean', 'wind_speed_10m_max', 'shortwave_radiation_sum', 'et0_fao_evapotranspiration']
dataForVisual = dataset[selectedCols] # Loading specipic data fro visuals
dataForVisual['datetime'] = pd.to_datetime(dataForVisual['datetime'])   # Extract date(string) into date-time format
dataForVisual.set_index('datetime').plot(figsize = (10, 5))       # Plot all numerial columns against time
plt.show()
```

Heatmap to visualize Correlation between numerical features

```python
import seaborn as sns
selectedCols1 = dataset.columns.to_list()
dataForCorr = dataset[selectedCols1].drop(columns = ['city_name', 'datetime', 'weather_code'])
plt.figure(figsize = (16, 14))
sns.heatmap(dataForCorr.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```
