Loaded Dataset

```python
import pandas as pd
dataset = pd.read_csv("F:\VS Code\Personal Projects\datasets\daily_data_combined_2020_to_YTD.csv\daily_data_combined_1980_to_1989.csv")
dataset.head()
```

Mean data store in place of missing values

```python
missing_data_cols = ['temperature_2m_mean', 'apparent_temperature_mean', 'sunshine_duration', 'precipitation_sum', 'rain_sum', 'snowfall_sum', 'wind_direction_10m_dominant', 'shortwave_radiation_sum', 'et0_fao_evapotranspiration']
dataset[missing_data_cols] = dataset[missing_data_cols].fillna(dataset[missing_data_cols].mean())
```

Droping unrequired columns

```python
dataset.drop(['sunrise', 'sunset'], axis = 1, inplace = True)
dataset.head()
```

Save to cleaned dataset file

```python
dataset.to_csv("cleaned_dataset.csv", index = False)
```
